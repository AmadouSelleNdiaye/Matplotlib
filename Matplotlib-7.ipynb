```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
satisfaction = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.grid(True)
plt.scatter(x,y,s=sizes, c=satisfaction, cmap='Greens',
            edgecolor='k',linewidth=1, alpha=0.75)
cbar= plt.colorbar()
cbar.set_label("Customers Satisfaction")
plt.tight_layout()
plt.show()
```

```python
data= pd.read_csv('2019-05-31-data.csv')
data.head()
```

```python
plt.scatter(data['view_count'],data['likes'],cmap='summer', c=data['ratio'],
            edgecolor='k',linewidth=1, alpha=0.75)

plt.xscale('log') # faire un zoom sur les donnes pour mieux comprendre
plt.yscale('log') 
cbar= plt.colorbar()
cbar.set_label("Likes/Dislikes Ratio")

plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.title('Trending Youtube Videos')

plt.grid(True)
plt.tight_layout()
plt.show()
```

# Notes 
- plt.xscale() or plt.yscale() : est utile lorsqu'il s'agit de donnes qui s'tendent sur plusieurs ordres de grandeur sur l'axe des x ou y, car elle permet de mieux visualiser la distribution et les tendances.
